**Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?**

MongoDB is non-relational database management system (DBMS) that falls under the category of NoSQL databases. 

MongoDB is usually preferred over SQL databases in multiple scenarios:

1. MongoDB's distributed architecture and horizontal scaling capabilities make it well-suited for handling large amounts of data and high-traffic applications.

2. MongoDB's support for indexing and aggregation pipelines makes it suitable for real-time analytics and complex querying, allowing you to extract insights from your data efficiently.

3.  If the data is unstructured or has varying structures across different records, MongoDB's document-oriented model can handle it more naturally compared to rigidly structured SQL databases.


**Q2. State and Explain the features of MongoDB.**

**Features of MongoDB**

1. MongoDB is designed to be highly scalable and supports horizontal scaling by distributing data across multiple servers or shards. 

2. MongoDB supports indexing on fields, including compound indexes and geospatial indexes, to optimize query performance.

3. MongoDB's aggregation framework allows for advanced data processing and analysis.

**Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.**

The sort() method in MongoDB is used to sort the results of a query in a specified order.

**Q7. Explain why delete_one(), delete_many(), and drop() is used.**

**delete_one()** method is used to delete a single document from a collection that matches a specific filter.

**delete_many()** method is used to delete multiple documents from a collection that match a specific filter.

**drop()** method is used to remove an entire collection from the database.
